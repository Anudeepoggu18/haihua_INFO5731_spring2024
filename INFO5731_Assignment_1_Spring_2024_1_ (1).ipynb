```python
import re

def calculate(expr):
    try:
        # Replace 'x' with '*' for multiplication and evaluate the expression
        result = eval(re.sub('x', '*', expr))
        return result
    except ZeroDivisionError:
        return "Error: Division by zero"
    except Exception as e:
        return f"Error: {e}"

def evaluate_with_parentheses(expr):
    # Use a loop to find innermost parentheses and evaluate them until no more parentheses are left
    while '(' in expr:
        # Use regular expression to find innermost parentheses
        match = re.search(r'\(([^()]+)\)', expr)
        if match:
            inner_expr = match.group(1)
            result = calculate(inner_expr)
            # Replace the innermost parentheses with the result
            expr = expr.replace(match.group(0), str(result))
        else:
            break
    return expr

def main():
    print(" operations: +, -, x , /")
    while True:
        user_input = input("Enter a mathematical expression (or 'exit' to quit): ")

        if user_input.lower() == 'exit':
            break

        # Evaluate expressions with parentheses
        result_with_parentheses = evaluate_with_parentheses(user_input)

        # Calculate the final result
        final_result = calculate(result_with_parentheses)

        print("Result:", final_result)

if __name__ == "__main__":
    main()
```
